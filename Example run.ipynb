# Hyper-parameters

Parameters for processing HateXplain can be found [here](https://github.com/upunaprosk/fair-pruning/blob/master/Parameters_description.md).


Parameters for removing layers:
```
  remove_layers: "6,7,8,9,10,11"
  freeze_layers: ""
  freeze_embeddings: False

```
Parameters are specified in config ```params.yml```.


````python
%%bash
echo "Default hyper-parameters"
echo "===== [ params.yml ] ====="
cat params.yml
echo "====="
````

Example of changing default config

````python
%%bash
cat <<__YML__ > params.yml
model: "bert-base-cased"
seed: 42
training:
  device: "gpu"
  batch_size: 16
  remove_layers: "6,7,8,9,10,11"
  freeze_layers: ""
  freeze_embeddings: False
  learning_rate: 2e-5
  epochs: 3
  auto_weights: True
  report_to: "wandb"
  train_att: True
  att_lambda: 1
  num_supervised_heads: 1
  supervised_layer_pos: 0
dataset:
  data_file: "Data/dataset.json"
  class_names: "Data/classes.npy"
  num_classes: 3
  max_length: 128
  include_special: False
  type_attention: "softmax"
  variance: 5.0
  decay: False
  window: 4.0
  alpha: 0.5
  p_value: 0.8
  method: "additive"
  normalized: False
logging: ""
__YML__
:
````

````python
import json
explainability_metrics_to_log = {}
with open(output_scores) as f:
    d = json.load(f)
for plaus in ["comprehensiveness", "sufficiency" ]:
    explainability_metrics_to_log[plaus] = d["classification_scores"][plaus]
for i,v in d["iou_scores"][0]['macro'].items():
    n = f"iou_scores_{i}"
    explainability_metrics_to_log[n] = v
for i,v in d["token_prf"]['instance_macro'].items():
    n = f"token_prf_{i}"
    explainability_metrics_to_log[n] = v
explainability_metrics_to_log["token_soft_metrics_auprc"] = d["token_soft_metrics"]["auprc"]
wandb.log(explainability_metrics_to_log)
````
